# Operation Data Analysys: High uncertainty
# Simulation with randome data generator
### run set at 1000, generate 1000 data points

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
import statistics

# Set the seed for reproducibility
random.seed(123)
run =1000

# Generate 10 data points for U with a mean of 23 and a deviation of 5
usage_data = [random.gauss(23, 5) for _ in range(run)]

# Calculate P for each data point
price_data = []
for u in usage_data:
    if u <= 20:
        price_data.append(160)
    else:
        price_data.append(160 + (u - 20) * 15)

# Calculate mean and standard deviation for U and P
mean_u = statistics.mean(usage_data)
std_dev_u = statistics.stdev(usage_data)

mean_p = statistics.mean(price_data)
std_dev_p = statistics.stdev(price_data)

# Create a dataframe with U and P data
data = {"U": usage_data, "P": price_data}
df = pd.DataFrame(data)

# Plot histograms for U and P
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(df['U'], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('U')
plt.ylabel('Frequency')
plt.title('Histogram of U\nMean: {:.2f}, Standard Deviation: {:.2f}'.format(mean_u, std_dev_u))

plt.subplot(1, 2, 2)
plt.hist(df['P'], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('P')
plt.ylabel('Frequency')
plt.title('Histogram of P\nMean: {:.2f}, Standard Deviation: {:.2f}'.format(mean_p, std_dev_p))

plt.tight_layout()
plt.show()
```
